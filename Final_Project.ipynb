```python
# for numerical column we prefer mean or median by finding the outliers
numerical_columns=['Avg VTAT','Avg CTAT','Booking Value','Ride Distance','Driver Ratings','Customer Rating']
for i in numerical_columns:
    df[i].plot(kind='box')
    plt.show()
```

```python
# The outliers are present for some columns then we prefer median 
median_prefer=['Booking Value','Driver Ratings','Customer Rating']
for i in median_prefer:
    print(f'The median of {i} is {df[i].median()}')
```

```python
# lets fill the null values using median for the columns :['Booking Value','Driver Ratings','Customer Rating']
for i in median_prefer:
    df[i]=df[i].fillna(df[i].median())
```

```python
# Now for remaining columns it has no outliers so we prefer mean 
mean_prefer=['Avg VTAT','Avg CTAT','Ride Distance']
for i in mean_prefer:
    print(f'The mean of {i} is {df[i].mean()}')
```

```python
# Lets fill the null values 
for i in mean_prefer:
    df[i]=df[i].fillna(df[i].mean())
```

```python
df[['Cancelled Rides by Customer','Cancelled Rides by Driver','Incomplete Rides']]=df[['Cancelled Rides by Customer',
                                                                                       'Cancelled Rides by Driver','Incomplete Rides']].fillna(0)
```

```python
df[['Reason for cancelling by Customer','Driver Cancellation Reason','Incomplete Rides Reason']]=df[['Reason for cancelling by Customer',
'Driver Cancellation Reason','Incomplete Rides Reason']].fillna('No Reason')
```

```python
df['Payment Method']=df['Payment Method'].fillna(df['Payment Method'].mode()[0])
```

```python
df.isnull().sum()
```

#### o	Convert date/time columns to appropriate formats.

```python
# Now o	Convert date and time columns to appropriate formats
df['Date']=pd.to_datetime(df['Date'])
df['Time']=pd.to_datetime(df['Time'],format='%H:%M:%S').dt.time
```
